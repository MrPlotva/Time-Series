```python
from matplotlib import pyplot as plt

epsilons = []
eps_left = 0.01
eps_right = 1
ROWS_CNT = 10
for i in range(ROWS_CNT):
    epsilons.append(eps_left + (eps_right - eps_left) / (ROWS_CNT - 1.0) * i)
comparisons = [30.15, 30.65, 29.633333333333333, 33.18333333333334, 36.516666666666666, 36.2, 30.3, 21.916666666666668, 18.78333333333333, 16.983333333333334]
plt.plot(epsilons, comparisons)
plt.title('Series similarity by epsilon')
plt.xlabel('Epsilon')
plt.ylabel('Similarity')
```

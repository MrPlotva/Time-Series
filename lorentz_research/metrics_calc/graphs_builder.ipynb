```python
import numpy as np

ROWS_CNT = 10
epsilons = []
eps_left = 0.01
eps_right = 1
for i in range(ROWS_CNT):
    epsilons.append(eps_left + (eps_right - eps_left) / (ROWS_CNT - 1.0) * i)


simple_metrics = np.load('../nnpp_for_tenk/NNPP_simple.npy',allow_pickle=True)
nnpp_rel = np.zeros((10, 4))

for index in range(10):
    loaded = np.load('../nnpp_for_tenk/NNPP_by_epsForTenK-' + str(index) + '.npy', allow_pickle=True)
    nnpp_rel[index] = simple_metrics / loaded

print(nnpp_rel)
```

```python
from matplotlib import pyplot as plt


def get_rel(index):
    rel = []
    for i in range(10):
        rel.append(nnpp_rel[i][index])
    return rel

plt.figure(figsize=(18, 12))
plt.plot(epsilons, get_rel(0))
plt.xlabel("Epsilons on hor 25", size=20)
plt.ylabel("NNPP rel", size=20)
plt.show()

plt.figure(figsize=(18, 12))
plt.plot(epsilons, get_rel(1))
plt.xlabel("Epsilons on hor 50", size=20)
plt.ylabel("NNPP rel", size=20)
plt.show()

plt.figure(figsize=(18, 12))
plt.plot(epsilons, get_rel(2))
plt.xlabel("Epsilons on hor 75", size=20)
plt.ylabel("NNPP rel", size=20)
plt.show()

plt.figure(figsize=(18, 12))
plt.plot(epsilons, get_rel(3))
plt.xlabel("Epsilons on hor 100", size=20)
plt.ylabel("NNPP rel", size=20)
plt.show()
```
